```python
# volume rendering

#!pip install ipyvolume

import ipyvolume
import yt
import numpy as np
```

```python
ds = yt.load('/Users/jillnaiman1/Downloads/enzo_tiny_cosmology/DD0027/DD0027')
```

```python
dd = ds.covering_grid(level=0, left_edge=[0,0,0], dims=ds.domain_dimensions)
```

```python
density3d = dd[('gas','density')].to_ndarray()
density3d, density3d.shape
```

```python
density3d = np.log10(density3d)
density3d = (density3d-density3d.min())/(density3d.max()-density3d.min())
```

```python
density3d
```

```python
ipyvolume.quickvolshow(density3d, level=[0.1,0.5,0.75], opacity=0.03, level_width=0.1)
```
