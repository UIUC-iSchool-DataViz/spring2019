```python
# Another idiosyncracy of pandas is its use of lambda functions
# lambda functions allow you to shorthand simple function definitions

# in this example x of 5 means run the lambda function on 5
# and the lambda function says return 5 after you add 10 to 5.
```

```python
x = lambda a : a + 10

x(5)
```

```python
# Pandas has an "apply" function that allows you to run a lambda function over each record in a dataframe
# This code will visit every date record, and return that record's year to a brand new year field
# Then it will do the same to the brand new hour field
```

```python
ufos["year"] = ufos["date"].apply(lambda a: a.year)
ufos["hour"] = ufos["date"].apply(lambda a: a.hour)
```

```python
# In this code, we're now grouping by unique pairs of year and shape
# counting how many records in each pair had valid entries by looking for the country in the records
#   (some years certain shapes are never reported so a year-shape pair might have no records)
```

```python
ufos.groupby(["year", "shape"])["country"].count()
```

```python
# and why not, we can plot this data

# you can see the data is quite noisey become some shapes are popular 
# but other shapes are reported between them each year
```

```python
ufos.groupby(["year", "shape"])["country"].count().plot()
```

```python
# Let's try plotting the prevalence of each shape as its own line

# We're creating a for loop that makes a dictionary key out of each shape
# and then adds a dataframe as the value
# where the dataframe is the total number of valid (country) records for each year

# the last line in the for loop then plots a line of that shape over time

# we can also draw a legend
# and resize using the rcParams attribute we learned last week
```

```python
shapes = {}
for shape, shapedata in ufos.groupby("shape"):
    shapes[shape] = shapedata.groupby("year")["country"].count()
    plt.plot(shapes[shape], label=shape)
plt.legend()
plt.rcParams["figure.figsize"] = [14,10]
```

```python
# now let's find the 5 most common shapes.

# We use value_counts to report this *series* and find the number of records in each shape category
```

```python
ufos["shape"].value_counts()
```

```python
# Then we truncate that to the top 5 count values
```

```python
ufos["shape"].value_counts()[:5]
```

```python
# Then we use this index attribute to get the metadata of the truncated series 
```

```python
ufos["shape"].value_counts()[:5].index
```

```python
# and then we extract a list of category names using the values attribute
```

```python
ufos["shape"].value_counts()[:5].index.values
```

```python
# let's store this list of category values in a variable
```

```python
most_common = ufos["shape"].value_counts()[:5].index.values
```

```python
# Now 

# Shapes[shape] is looking back into the dictionary we created earlier
# which stored the number of sightings per year for each shape

# When we divide that by the total sightings per year
# we see what percentage of a year's sightings belonged to each shape of UFO

# So you can see that a large number of early sightings were circles
# and there's been a recent growth in just light and drop off in triangles

# (side note, notice that the plot rcParams figsize is maintained from the last plot)
```

```python
total_sightings = ufos.groupby("year").count()
for shape in most_common:
    plt.plot(shapes[shape]/total_sightings["country"], label=shape)
plt.legend()
```

```python
# this could be clearer as a stacked line plot
# which would show each line as parts of a whole
```

```python
total_sightings = ufos.groupby("year").count()
runningTotal = shapedata.groupby("year")["country"].apply(lambda a: 0)
for shape in most_common:
    runningTotal = runningTotal+shapes[shape]
    plt.plot(runningTotal/total_sightings["country"], label=shape)
plt.legend()
```

```python
# In this next example, we're creating a grid of line graphs that are related to each other by row and column

# we're comparing each shape's prevalence with respect to one other shape
# diagonal plots are inverses of each other
```

```python
plt.figure(dpi = 300)
multi_shapes = ["disk", "light", "fireball", "chevron"]
for i, shape1 in enumerate(multi_shapes):
    for j, shape2 in enumerate(multi_shapes):
        plt.subplot(len(multi_shapes), len(multi_shapes), i * len(multi_shapes) + j + 1)
        if shape1 == shape2:
            plt.text(0.25, 0.5, shape1)
            plt.xticks([])
            plt.yticks([])
            continue
        plt.plot(shapes[shape1] / (shapes[shape2] + shapes[shape1]))
        plt.xlim(1940, 2014)
        plt.ylim(0, 1.0)
        plt.xticks([])
        plt.yticks([])
```
