```python
# so now we're going back to the beginning, and let's just print everything.
# we're going to use a special kind of iterator defined by the CSV library.
```

```python
f.seek(0)
for record in csv.reader(f):
    print(record)
```

```python
# You can see that this automatically formats every line in the CSV file as a list
# You can also see that the first line of the file is different - it's not a data entry, it's a header.
# Let's try parsing the header.
```

```python
# first, we'll go back to the start of the file
f.seek(0)
# then we'll store our CSV iterator with a variable called reader
reader = csv.reader(f)
# then we'll read in ONLY THE FIRST LINE of the CSV file with the next function
header = next(reader)
# we initialize a dictionary
data = {}
# and we loop through each of the data field names in the header, initializing them as dictionary keys
for col in header:
    data[col] = []

data
```

```python
# we also have access to zip, which can combine two separate lists into one list of pairs
```

```python
a = ["Arcanine", "is", "my", "favorite"]
b = [9, 4, 1, 9]
for x in zip(a, b):
    print(x)
```

```python
# so now let's fill in our dictionary
```

```python
# all of this is the same
f.seek(0)
reader = csv.reader(f)
header = next(reader)
data = {}
for col in header:
    data[col] = []

# now we're going to set up a loop through each line of the CSV file with the reader iterator
for row in reader:
    # and we're going to use zip to combine each value in a row with its data field name
    # which we previously stored in the list called header
    # ...
    # then we're going to go through each pair and add it to the dictionary
    # using the data field name as the key, and the information from the row as the value.
    for col, val in zip(header, row):
        data[col].append(val)
```

```python
# so let's say I know the pokemon Charizard, but I want to know what pokemon evolves into it
```

```python
# first I look through the list of all pokemon names to try to find charizard
for i, name in enumerate(data['identifier']):
    if 'charizard' in name:
        # and when I find it, I use the index to look up the pokemon number of the species it comes from
        # I subtract 1 because python starts counting at zero but pokemon start at 1
        younger = int(data['evolves_from_species_id'][i]) - 1
        # and then I print the name of the pokemon at that younger index position we just got
        print(data['identifier'][younger])
```

```python
# now I'm going to use a new datatype called a Counter.
# to do this I need to import a library called collections.
# this library extends python's datatypes to do efficient processing of large datasets.
```

```python
import collections
```

```python
# I'm going to create a Counter object that rapidly counts the items in a dictionary
# This creates a new dictionary of each value as a key and it's count as the only value
```

```python
collections.Counter(data['is_baby'])
```

```python
# so we can see there are 18 baby pokemon and 789 adult pokemon(?)
# 
# Now let's try to pull out a subset of the data based on a field
```

```python
# I'm going to find all the green pokemon
# so first I initialize a dictionary just for the green pokemon
green = {}
for field in header:
    green[field] = []

# then I iterate through the dictionary looking for the correct color
for i, color in enumerate(data['color_id']):
    if color == 'green':
        # and when I find it, I add the key,value pairs for every data field into my new green-only dictionary
        for key in data:
            green[key].append(data[key][i])
```

```python
# so now we can see how many green pokemon there are with the collections Counter again
```

```python
collections.Counter(green['color_id'])
```
