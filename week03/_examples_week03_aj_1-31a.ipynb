```python
# a lightweight example here shows how this tool works

# first you create a function
# then you feed the possible function arguments to the ipywidgets' interact function
# this @ sign indicates a decorator function - we place it before the newly defined function
# to notify you that it is modifying the function that follows

# You can see that when we change the current dropdown item
# the function re-evaluates and prints the current option below
```

```python
@ipywidgets.interact(myList=["option 1","option 2","option 3"])
def g(myList):
    return (myList)
```

```python
# now let's make a dropdown of the available styles so we can change them interactively!
# we can give our interactive plot function any name. Let's call it "make plot"

# we give it an argument "style" and then override that in the the decorator function's argument
# and we set it equal to the list of possible plot styles
```

```python
@ipywidgets.interact(style = plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        plt.plot(stats["max"], linewidth = 0.5, marker='.', label = "Maximum Acquisition")
        plt.plot(stats["min"], linewidth = 0.5, marker='.', label = "Minimum Acquisition")
        plt.fill_between(stats.index, stats["min"], stats["max"], color = "#dddeee")
        plt.xlabel("Year")
        plt.ylabel("Square Footage")
        plt.yscale("log")
        plt.legend()
```

```python
# let's say we want this line to be a function of the size of the plot IMAGE 
# instead of using values specific to the plot

# to do that, we need to work with TRANSFORM data, which we can access with this library
```

```python
import matplotlib.transforms as mpt
```

```python
# now we're going to create a transform

# First we retrieve our "AXES" object. This is a very confusing term that basically just means the plotting canvas.
# It is not to be confused with an AXIS, which is a line along one side of the graph.

# plt.gca means "Get Current Axes"

# Then we create a transformation that we can use to place the black line.
# the blended_transform_factory function returns an X/Y transform derived from the X argument and the Y argument
# transData is the transformation across the DATA range
#     this means the x-axis of our line will be based on data
# transAxes is the transformation across the AXES canvas object
#     this means the y-axis of our line will be based on the canvas size

# Then finally we add the transformation as an argument to the line plot.
# Because the line is in canvas space now, it only has to go from 0 to 1
#     -- basically to 100% of the height of the canvas
```

```python
plt.plot(stats["max"], marker='.', linewidth = 1.0, label = "Max")
plt.plot(stats["min"], marker='.', linewidth = 1.0, label = "Min")
plt.fill_between(stats.index, stats["min"], stats["max"], color = "#aaaaaa")
plt.xlabel("Year")
plt.ylabel("Square Footage")
plt.yscale("log")

ax = plt.gca()
new_transform = mpt.blended_transform_factory(ax.transData, ax.transAxes)
plt.plot([1818, 1818], [0.0, 1.0], color = 'black', linewidth = 2.0, transform = new_transform)

plt.legend()
```

```python
# for the next example, we're going to use the rolling function in pandas
# rolling creates groups of adjacent data
# the size of the groups is the argument to the rolling function

# so for instance, if we want to take the total square footage acquired per year
# and then find the sum every year of the past three years, we might do this.

# You can see the first couple entries are Not a Number, 
# because there wasn't enough data in those years to calculate a three year sum
```

```python
df.groupby("Year Acquired")["Square Footage"].sum().rolling(3).sum()
```

```python
# so let's try an interactive visualization that shows rolling averages

# first we define a function
```

```python
# then we divide the data into groups of adjacent entries
# because we want the size of the window to be the interactive value, we make it the argument to the function
# I'm going to find the mean of each window, and then calculate a mean of means
# so we'll get an approximate average of the years in the window
```

```python
def make_plot(window):
    df2 = stats.rolling(window = window)["mean"].mean()
```

```python
# now let's add the decorator function back
# we're going to set the default rolling window to a size of 1, so just individual records to start
# these other two arguments in the window tuple will be the maximum and minimum values the interactive slider allows
```

```python
@ipywidgets.interact(window = (1, 12, 1))
def make_plot(window):
    df2 = stats.rolling(window = window)["mean"].mean()
```

```python
# Now let's add the rest of the plotting code in
# First we want to plot the average square footage acquired over time
# Then we want to plot the average square footage acquired in windowed increments, using our new windowed dataframe
# and we'll use a log scale again so large numbers are easier to analyze

# when we run this, you can see there's now a slider to adjust the window size
# you can see that the Not A Numbers at the beginning of the windows don't plot anything
# and you can see the plot is telling us more about the general trend of square footage acquisitions
# than the original mean sampled every year.
```

```python
@ipywidgets.interact(window = (1, 12, 1))
def make_plot(window):
    df2 = stats.rolling(window = window)["mean"].mean()
    plt.plot(stats["mean"], marker = '.', linewidth = 0.5, alpha = 0.5)
    plt.plot(df2, marker = '.', linewidth = 1.5)
    plt.yscale("log")
```
