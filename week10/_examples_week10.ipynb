```python
import glob
```

```python
fns = sorted(glob.glob("/Users/jillnaiman1/spring2019online/week10/data/names/yob*.txt"))
```

```python
data = []
for fn in fns:
    year = int(fn[-8:-4])
    df = pd.read_csv(fn, header=None, names = ["name", "sex", "count"])
    df["year"] = year
    data.append(df)
```

```python
all_names = pd.concat(data)
```

```python
most_common = []
for year, df in all_names.groupby("year"):
    imax = df["count"].idxmax()
    most_common.append((year, df["name"].loc[imax], df["count"].max()))
most_common = pd.DataFrame(most_common, columns = ["year", "name", "count"])
most_common.set_index("year", inplace=True)
```
